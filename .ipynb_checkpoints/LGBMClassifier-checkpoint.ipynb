```python
import optuna
from lightgbm import LGBMClassifier
import lightgbm as lgb
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
import folium
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.utils import resample
import shap
```

```python
df = pd.read_csv("sungdong_preprocessed.csv")
df.head()
```

```python
# # 이진 분류로 바꿔서 진행
# df['istow'] = (df['tow_count'] >= 1).astype(int)
```

```python
# 지역 기반 분할
train_df = df[df['SIG_KOR_NM'].isin(['강남구', '서초구', '성동구'])]
val_df   = df[df['SIG_KOR_NM'] == '마포구']
test_df  = df[df['SIG_KOR_NM'] == '송파구']
```

```python
df_majority = train_df[train_df['tow_count'] == 0]
df_minority = train_df[train_df['tow_count'] != 0]

n_minority = len(df_minority)
n_majority_target = int(n_minority)

# 언더샘플링
df_majority_downsampled = resample(df_majority,
                                   replace=False,
                                   n_samples=n_majority_target,
                                   random_state=42)

# 병합
df_balanced = pd.concat([df_majority_downsampled, df_minority])

# 셔플
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

train_df = df_balanced.copy()

# 비율 확인
print(df_balanced['tow_count'].value_counts(normalize=True))
```

### 클래스 구간 분류

```python
df_nonzero = train_df[train_df['tow_count'] > 0].copy()

# `tow_count` 값으로 정렬
df_nonzero = df_nonzero.sort_values(by='tow_count')

# 3등분 비율 계산
total_nonzero = len(df_nonzero)

# 각 구간 끝 인덱스 구하기
group1_end = int(total_nonzero * 1/3)  # 1/3 비율
group2_end = int(total_nonzero * 2/3)  # 2/3 비율

# 각 구간 끝값 (경계)
group1_end_value = df_nonzero.iloc[group1_end - 1]['tow_count']  # 첫 번째 구간 끝값
group2_end_value = df_nonzero.iloc[group2_end - 1]['tow_count']  # 두 번째 구간 끝값
group3_end_value = df_nonzero.iloc[total_nonzero - 1]['tow_count']  # 세 번째 구간 끝값

print(f"첫 번째 구간 끝값: {group1_end_value}")
print(f"두 번째 구간 끝값: {group2_end_value}")
print(f"세 번째 구간 끝값: {group3_end_value}\n")


train_df['risk_level'] = train_df['tow_count'].apply(
    lambda x: 0 if x == 0 else (1 if x == 1 else (2 if x < 4 else 3))
)
val_df['risk_level'] = val_df['tow_count'].apply(
    lambda x: 0 if x == 0 else (1 if x == 1 else (2 if x < 4 else 3))
)
test_df['risk_level'] = test_df['tow_count'].apply(
    lambda x: 0 if x == 0 else (1 if x == 1 else (2 if x < 4 else 3))
)
# 결과 확인
print(train_df[['tow_count', 'risk_level']].tail())
```

### 훈련/검증/테스트 셋 분류

```python
train_df.columns
```

```python
features = train_df.columns[3:-3]
target = 'risk_level'

# 학습/검증/테스트 데이터 준비
X_train, y_train = train_df[features], train_df[target]
X_val, y_val     = val_df[features], val_df[target]
X_test, y_test   = test_df[features], test_df[target]
```

```python
def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 300, 1000),  # 트리 개수
        "max_depth": trial.suggest_int("max_depth", 4, 16),  # 트리의 최대 깊이
        "num_leaves": trial.suggest_int("num_leaves", 20, 200),  # 리프 노드 개수
        "min_child_samples": trial.suggest_int("min_child_samples", 5, 100),  # 리프 노드가 가져야 할 최소 샘플 수
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "learning_rate": trial.suggest_float("learning_rate", 1e-4, 0.2, log=True),
        "class_weight": 'balanced',
        "random_state": 42,
        "n_jobs": -1,
        "verbose": -1
    }

    model = LGBMClassifier(**params)
    model.fit(X_train, y_train)  # 모델 학습

    y_pred = model.predict(X_val)  # 예측
    accuracy = accuracy_score(y_val, y_pred)  # 정확도 계산
    
    return accuracy 
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

print("Best params:", study.best_params)
print("Best accuracy:", study.best_value)
```

```python
best_model = LGBMClassifier(**study.best_params, random_state=42, n_jobs=-1)
best_model.fit(X_train, y_train)

# 검증 데이터로 예측하고 Accuracy 계산
val_preds = best_model.predict(X_val)
accuracy = accuracy_score(y_val, val_preds)
print(f"Validation Accuracy: {accuracy:.4f}")

# 테스트 데이터로 예측하고 Accuracy 계산
test_preds = best_model.predict(X_test)
accuracy = accuracy_score(y_test, test_preds)
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
val_df = val_df.copy()
val_df['pred'] = val_preds

m = folium.Map(location=[37.5056, 127.1167], zoom_start=13)

for _, row in val_df.iterrows():
    lat = row['centroid_lat']
    lon = row['centroid_lon']
    pred = row['pred']
    actual = row['risk_level']
    
    error = abs(pred - actual) # 예측 오차
    if error == 3:
        color = 'red'
    elif error == 2:
        color = 'yellow'
    else:
        color = 'green'


    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        fill=True,
        color = color,
        fill_color=color,
        fill_opacity=0.7,
        popup=folium.Popup(f"예측: {pred:.2f}<br>실제: {actual}", max_width=150)
    ).add_to(m)

m
```
